```python
X =  dataset_new[["Years of Experience"]]
y = dataset_new[["Salary"]]

# Create the data using scatter plot

import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
```
